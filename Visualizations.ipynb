```python
import numpy as np
import matplotlib.pyplot as plt
```

# Orifice Meter DP
The first visualization is used to show that the ideal operating range of differential pressures is 10-100 in H2O.  This is because values lower than 10 mean a small change in DP corresponds to a relatively larger change in flow rate, creating measurement uncertainty, and values higher than 100 mean that for a small increase in flow, there needs to be a large increase in differential pressure, resulting in greater permanent pressure loss and unreasonable operating conditions.

```python
# Generate values for x
x = np.linspace(0, 120, 400)
# Calculate the square root of x
y = 2*np.sqrt(x)

# Plot curve
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Flow Rate (MSCFD)')
plt.fill_betweenx(y*1.2, 10, 100, color='lightgreen', alpha=0.5)
plt.text(50, 8, 'Optimal Operating DP', horizontalalignment='center', verticalalignment='center', bbox=dict(facecolor='white', alpha=0.5))
plt.title('Flow Rate vs. DP of Example Orifice Meter')
plt.xlabel('Operating Differential Pressure ($in \: H_2O$)')
plt.ylabel('Flow Rate (MSCFD)')
plt.ylim(0,22.5)
plt.grid(True)
plt.legend()
plt.show()
```
